# 2. Generate Arrays of Different Dimensions.

Generate five NumPy arrays using np.arange():

0D array (a single value)

1D array (e.g., 0 to 4)

2D array (reshape a 1D array into 2x3)
 
4D array (bonus!)




```python
import numpy as np
a0=np.array(1)
a1=np.arange(0,4)
a2=np.arange(6).reshape(2,3)
a3=np.arange(8).reshape(2,2,2)
a4=np.arange(16).reshape(2,2,2,2)
print(a0,a0.ndim,"D")
print()
print(a1,a1.ndim,"D")
print()
print(a2,a2.ndim,"D")
print()
print(a3,a3.ndim,"D")
print()
print(a4,a4.ndim,"D")
```

# 3. Print the datatype, shape, size, and dimension of each of the above arrays

Use .dtype, .shape, .size, and .ndim.


```python
print(f"{a0} dtype-{a0.dtype} , shape-{a0.shape} , size-{a0.size} , Dimension-{a0.ndim}")
print()
print(f"{a1} dtype-{a1.dtype} , shape-{a1.shape} , size-{a1.size} , Dimension-{a1.ndim}")
print()
print(f"{a2} dtype-{a2.dtype} , shape-{a2.shape} , size-{a2.size} , Dimension-{a2.ndim}")
print()
print(f"{a3} dtype-{a3.dtype} , shape-{a3.shape} , size-{a3.size} , Dimension-{a3.ndim}")
print()
print(f"{a4} dtype-{a4.dtype} , shape-{a4.shape} , size-{a4.size} , Dimension-{a4.ndim}")
```
