# 9. Write a program to find common values between two arrays.
Use the lists given below:

prices_monday = [10, 30, 50, 60, 90]

prices_tuesday = [50, 20, 80, 125, 10, 70]


```python
import numpy as np
prices_monday = [10, 30, 50, 60, 90]
prices_tuesday = [50, 20, 80, 125, 10, 70]
a1=np.array(prices_monday)
a2=np.array(prices_tuesday)
comm=np.intersect1d(a1,a2)
print(comm)
```

# 10. Write a program to find BMI of the patients whose height and weight is provided in the list given below
Use the lists given below:

list of weights (in kg)
weights = [57, 69, 54, 65 ,71, 62, 56]
heights = [1.59, 1.75, 1.66, 1.74, 1.64, 1.72, 1.53] 

```python
import numpy as np
weights = [57, 69, 54, 65, 71, 62, 56]
heights = [1.59, 1.75, 1.66, 1.74, 1.64, 1.72, 1.53]
w=np.array(weights)
h=np.array(heights)
bmi=w/(h**2)
print("BMI of patients:",bmi)
```

# 19. Define a for loop to apply the function (2x-1)/(x+5) on every element of an array
Use the list given below:

numbers = [6, 5, 3, 15]


```python
import numpy as np
numbers=[6,5,3,15]
arr=np.array(numbers)
for i in np.nditer(arr):
    res=(2*i - 1) / (i + 5)
    print(res)
```
