ARRAY PROPERTIES AND ATTRIBUTES

```python
arr9=np.array([[1,2,3],
              [4,5,6]])
print(arr9)

print(arr9.shape) # no. rows and columns
print(arr9.size) #total no. of elements
print(arr9.ndim) # 1D OR 2D OR 3D
print(arr9.dtype) #data type

str_arr=arr9.astype(str)
print(str_arr) #convt to string type
```

ARRAY MATHS OPERATIONS

```python
arr10=np.array([10,50,40])
print(np.add(arr10,5))
print(np.subtract(arr10,5))
print(np.multiply(arr10,5))
print(np.divide(arr10,5))
print(np.floor_divide(arr10,4))
print(np.power(arr10,5))
print(np.mod(arr10,6))
```

ARRAY STATISTICAL OR AGGREGATE FUNCTIONS

```python
arr11=np.array([[1,2,3],
               [4,5,6]])
print(arr11)
print(np.sum(arr11,axis=1))
print(np.mean(arr11,axis=0))
print(np.min(arr11))
print(np.max(arr11))
print(np.var(arr11))
print(np.std(arr11))
print(arr11.argmin())
print(arr11.argmax())
```

ARRAY INDEXING AND SLICING

```python
arr12=np.array([10,20,30,40,50,60,70,80,90])
print(arr12)
print(arr12[0])
print(arr12[4])

print(arr12[2:8:3])
print(arr12[::-1])
```

```python
#fancy indexing-selecting multi elements at once
print(arr12[[0, 4, 8]])
```

```python
# boolean masking or filtering -filtering values based on some condition
print(arr12[arr12>25])
```

ARRAY RESHAPING AND MANIPULATION

```python
arr13=np.array([10,20,30,40,50,60])
res_array=arr13.reshape(2,3)
print(res_array)
```

```python
print(arr11.ravel()) #view - original array modified
print(arr11.flatten()) #copy - original array uneffected
```

HANDLING MISSING VALUES

```python
array=np.array([1,2,np.nan,4])
print(array)
print(np.isnan(array))
print(np.nan_to_num(array,nan=90))
```
