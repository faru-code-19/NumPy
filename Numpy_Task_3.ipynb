# 1. Create two 1D arrays

One containing multiples of 3 in variable (a)

One containing multiples of 5 in variable (b)


```python
import numpy as np
a=np.arange(3,30,3)
b=np.arange(5,50,5)
print(a)
print(b)
```

# 3. Statistical Summary
For the arrays a and b (created in previous questions or create a new one):

Find and print the Mean (average)

Find and print the Median

Find and print the Mode

Find and print the Range (Max - Min)

Find and print the Standard Deviation

```python
arr=np.arange(1,21)
print(arr)
print("Mean",np.mean(arr))
print("Median",np.median(arr))
print("Range",np.max(arr)-np.min(arr))
print("Std",np.std(arr))
print("Variance",np.var(arr))
print("Sum",np.sum(arr))
```

# 4. Use flip() to Reverse

Create a 1D NumPy array with values from 10 to 50 using np.arange(). Reverse the array using np.flip().


```python
ar=np.arange(10,51)
print(ar)
rev=np.flip(ar)
print(rev)
```

# 5. Reverse Without flip()
Achieve the same reversed array without using np.flip().


```python
ar=np.arange(10,51)
print(ar)
r=ar[::-1]
print(r)
```

# 7. Conditional Replacement
Replace all even numbers in the reversed array (or a newly created array) with -1 and print the modified array.


```python
ar=np.arange(10,51)
r=ar[::-1]
print(r)
r[r%2==0]=-1
print(r)
```

# 10. Array Iteration

Create a 2D NumPy array of shape (2, 3) with any integers.

Iterate through the array using a basic for loop.

Then, iterate using np.nditer().


```python
import numpy as np
a=np.random.randint(1,11,(2,3))
print(a,a.ndim,"D")
```

```python
for i in a:
    for j in i:
        print(j)
```

```python
for i in np.nditer(a):
    print(i)
```

# 11. Operations on Random Float Arrays
Generate two 1D arrays (length 5 each) with random float numbers between 1 and 10.

Perform element-wise operations: +, -, *, /, %, //.

Apply mathematical functions: np.sqrt(), np.exp(), np.sin(), np.log().


```python
import numpy as np

a=np.random.uniform(1,10,5)
b=np.random.uniform(1,10,5)
print("a:",a)
print("b:",b)

print("\nAddition",a+b)
print("\nSubtraction",a-b)
print("\nMultiplication",a*b)
print("\nDivision",a/b)
print("\nModulus",a%b)
print("\nFloor Division",a//b)
print("\nSquare Root of a",np.sqrt(a))
print("\nExponential of a",np.exp(a))
print("\nSine of a",np.sin(a))
print("\nLog of a",np.log(a))
```
